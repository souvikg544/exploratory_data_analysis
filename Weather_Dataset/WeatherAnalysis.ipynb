```python
time = pd.to_datetime(weather_dataset['Date/Time'])
temp = weather_dataset['Temp_C']
humidity = weather_dataset['Rel Hum_%']
windSpeed = weather_dataset['Wind Speed_km/h']
visibility = weather_dataset['Visibility_km']
pressure = weather_dataset['Press_kPa']
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.plot(time,temp)
rcParams['figure.figsize'] = 15,10
plt.grid(True,color='k',linestyle=':')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Time')
ax.set_ylabel('Humidity')
ax.plot(time,humidity)
rcParams['figure.figsize'] = 15,10
plt.grid(True,color='k',linestyle=':')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Time')
ax.set_ylabel('WindSpeed')
ax.plot(time,windSpeed)
rcParams['figure.figsize'] = 15,10
plt.grid(True,color='k',linestyle=':')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.set_xlabel('Time')
ax.set_ylabel('Pressure')
ax.plot(time,pressure)
rcParams['figure.figsize'] = 15,10
plt.grid(True,color='k',linestyle=':')
plt.show()
```

## Finding out How many times Team won the match on winning the Toss

```python
new_df=init_df[(init_df["toss_won"]==init_df["winner"])]
```

```python
total_match=len(init_df)
wrong_decision=total_match-len(new_df)
right_decision=len(new_df)
```

```python
labels = ['Right Decision', 'Wrong Decision']
sizes = [right_decision, wrong_decision]
colors = ['lightblue', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Toss Decision affecting Wins')
plt.axis('equal')
plt.show()
```

## Most Wrong decisions taken by a Team after winning the toss

```python
gt=len(init_df[(init_df["toss_won"]=="Gujarat Titans")&(init_df["winner"]!="Gujarat Titans")])
csk=len(init_df[(init_df["toss_won"]=="Chennai Super Kings")&(init_df["winner"]!="Chennai Super Kings")])
rcb=len(init_df[(init_df["toss_won"]=="Royal Challengers Bangalore")&(init_df["winner"]!="Royal Challengers Bangalore")])
mi=len(init_df[(init_df["toss_won"]=="Mumbai Indians")&(init_df["winner"]!="Mumbai Indians")])
kkr=len(init_df[(init_df["toss_won"]=="Kolkata Knight Riders")&(init_df["winner"]!="Kolkata Knight Riders")])
pk=len(init_df[(init_df["toss_won"]=="Punjab Kings")&(init_df["winner"]!="Punjab Kings")])
lsg=len(init_df[(init_df["toss_won"]=="Lucknow Super Giants")&(init_df["winner"]!="Lucknow Super Giants")])
rr=len(init_df[(init_df["toss_won"]=="Rajasthan Royals")&(init_df["winner"]!="Rajasthan Royals")])
dc=len(init_df[(init_df["toss_won"]=="Delhi Capitals")&(init_df["winner"]!="Delhi Capitals")])
srh=len(init_df[(init_df["toss_won"]=="Sunrisers Hyderabad")&(init_df["winner"]!="Sunrisers Hyderabad")])
```

```python
gt_p=len(init_df[(init_df["toss_won"]=="Gujarat Titans")])
csk_p=len(init_df[(init_df["toss_won"]=="Chennai Super Kings")])
rcb_p=len(init_df[(init_df["toss_won"]=="Royal Challengers Bangalore")])
mi_p=len(init_df[(init_df["toss_won"]=="Mumbai Indians")])
kkr_p=len(init_df[(init_df["toss_won"]=="Kolkata Knight Riders")])
pk_p=len(init_df[(init_df["toss_won"]=="Punjab Kings")])
lsg_p=len(init_df[(init_df["toss_won"]=="Lucknow Super Giants")])
rr_p=len(init_df[(init_df["toss_won"]=="Rajasthan Royals")])
dc_p=len(init_df[(init_df["toss_won"]=="Delhi Capitals")])
srh_p=len(init_df[(init_df["toss_won"]=="Sunrisers Hyderabad")])
```

```python
gt_s=(gt/gt_p)*100
csk_s=(csk/csk_p)*100
rcb_s=(rcb/rcb_p)*100
mi_s=(mi/mi_p)*100
kkr_s=(kkr/kkr_p)*100
pk_s=(pk/pk_p)*100
lsg_s=(lsg/lsg_p)*100
rr_s=(rr/rr_p)*100
dc_s=(dc/dc_p)*100
srh_s=(srh/srh_p)*100
```

```python
teams = ['dc', 'srh', 'gt', 'csk', 'rcb', 'mi', 'kkr', 'pk', 'lsg', 'rr']
values = [dc_s, srh_s, gt_s, csk_s, rcb_s, mi_s, kkr_s, pk_s, lsg_s, rr_s]
plt.bar(teams, values)
plt.xlabel('Teams')
plt.ylabel('Percentage of times Toss won yet match lost')
plt.title('Comparing the percentage of times team won the toss yet lost the match')
plt.xticks(rotation=45)
plt.show()
```

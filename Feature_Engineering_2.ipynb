# Data Prediction

```python
#Now that we know that Property price is influenced Property location, Let's build 
#a simple Linear Regression Model to predict Prices of Property
```

```python
from sklearn.preprocessing import LabelEncoder
 
# Creating a instance of label Encoder.
le = LabelEncoder()
 
# Using .fit_transform function to fit label
# encoder and return encoded label
label = le.fit_transform(df['Location'])
 
# printing label
label

df.drop("Location", axis=1, inplace=True)
 
# Appending the array to our dataFrame
# with column name 'Purchased'
df["Location"] = label
 
```

```python
#we will select Property Location as a feature. 
feature = ["Location"]
```

```python
X_train = df[feature]
```

```python
X_train.head()
```

```python
#our target would be the price of the property
target = ["Price"]
```

```python
Y_train = df[target]
```

```python
Y_train.head()
```

```python
#Baselining is the first stage of model construction. Asking yourself this question will help you determine 
#how well your model is performing. Analyzing how a "dumb" model would perform on the same data is one method 
#to approach this problem. This model, which some people also refer to as naive or baseline, always produces 
#only one forecast; in this case, it predicts the same price regardless of the size of an apartment. So let's 
#start by determining what the prediction of our base model should be.
```

```python
#Calculate the mean of your target vector y_train and assign it to the variable y_mean.

Y_mean = Y_train.mean()
```

```python
#Now that we have the one prediction that our dumb model will always make, 
#we need to generate a list that repeats the prediction for every observation in our dataset.

y_pred_baseline = [Y_mean] * len(Y_train)
```

```python
#So how does our baseline model perform? 

plt.scatter(X_train, Y_train)
plt.xlabel("Location")
plt.ylabel("Price")
plt.title("Location vs Price")
plt.legend();
```

```python
#Looking at this visualization, it seems like our baseline model 
#doesn't really follow the trend in the data. But, as a data scientist, 
#you can't depend only on a subjective plot to evaluate a model. 
#You need an exact, mathematically calculate performance metric. 
#There are lots of performance metrics, but the one we'll use here is the mean absolute error.
mae_baseline = mean_absolute_error(Y_train, y_pred_baseline)

print("Mean apt price: ", round(Y_mean,2))
print("Baseline MAE: ", round(mae_baseline,2))
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#The final step is to evaluate our model. In order to do that, 
#we'll start by seeing how well it performs when making predictions 
#for data that it saw during training. So let's have it predict the price for the houses in our training set.

Y_pred_training = model.predict(X_train)
```

```python
mae_training = mean_absolute_error(Y_train, Y_pred_training)
print("Training MAE: ", round(mae_training,2))
```

```python
# Importing the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading in the dataset
df = pd.read_csv('data/Elections/winners/2019.csv')
df.head()
```

```python
# Number of different parties contesting elections along with contesting from how many constituencies
parties_contesting = pd.DataFrame(np.unique(df['Party'] ,return_counts=True)).T
parties_contesting.columns = ['Party' ,'No. of Constituencies won']

# Sorting by the largest winners
parties_contesting.sort_values('No. of Constituencies won' ,ascending=False)
```

#### So a total of 37 different parties contested the 2019 elections.

### Plotting the above dataframe using a barplot.

### *Note : Since party 'BJP' has a very very high constituency count compared to rest(301) ,we will not plot it on the barplot as it will disturb the scale of whole plot*

```python
# Plotting the number of seats won by each party except bjp

plt.figure(figsize=(18,9))
plt.xticks(color = 'red' ,size = 12 ,rotation = 85)
plt.yticks(color = 'red' ,size = 13)

ax = sns.barplot(x = parties_contesting['Party'].drop(8) ,y = parties_contesting['No. of Constituencies won'].drop(8).astype('int16'))


plt.xlabel('PARTY' ,color = 'violet',size = 15)
plt.ylabel('NO. OF  CONSTITUENCIES  WON' ,color = 'violet' ,size = 15)

plt.grid(b = True ,
         linewidth = 0.7,
         linestyle = '--',
         color = 'gray')

for bar in ax.patches:
       ax.annotate( s = str(bar.get_height()) , # the text to put
                    xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                    ha='center',
                    va='center',
                    size=14,
                    xytext=(0, 10),
                    textcoords='offset points')
```

```python
# Checking the median of numerical columns party-wise
df.groupby(['Party']).median()
```

```python
# Plotting the median criminal cases per winner for each party in a bar plot

plt.figure(figsize=(18,9))
plt.xticks(color = 'red' ,size = 12 ,rotation = 85)
plt.yticks(color = 'red' ,size = 13)

ax = sns.barplot(x = df.groupby(['Party']).median().index ,y=df.groupby(['Party']).median()['Criminal Cases'])


plt.xlabel('PARTY' ,color = 'violet',size = 15)
plt.ylabel('MEDIAN CRIME CASES PER WINNER' ,color = 'violet' ,size = 15)

plt.grid(b = True ,
         linewidth = 0.7,
         linestyle = '--',
         color = 'gray')

for bar in ax.patches:
       ax.annotate( s = str(bar.get_height()) , # the text to put
                    xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                    ha='center',
                    va='center',
                    size=14,
                    xytext=(0, 10),
                    textcoords='offset points')
```

```python
# Plotting the median Total Assets per winner for each party in a bar plot

plt.figure(figsize=(18,9))
plt.xticks(color = 'red' ,size = 12 ,rotation = 85)
plt.yticks(color = 'red' ,size = 13)

ax = sns.barplot(x = df.groupby(['Party']).median().index ,y=df.groupby(['Party']).median()['Total Assets'])


plt.xlabel('PARTY' ,color = 'violet',size = 15)
plt.ylabel('MEDIAN TOTAL ASSETS PER WINNER' ,color = 'violet' ,size = 15)

plt.grid(b = True ,
         linewidth = 0.7,
         linestyle = '--',
         color = 'gray')
```

```python
# Plotting the distributions of educational qualification of all winners
plt.figure(figsize=(20,7))
plt.xticks(color = 'red' ,size = 12)
plt.yticks(color = 'red' ,size = 13)

ax = sns.countplot(x=df['Education'])

plt.xlabel('EDUCATIONAL BACKGROUND' ,color = 'violet',size = 15)
plt.ylabel('NO OF CONTESTANTS' ,color = 'violet' ,size = 15)

plt.grid(b = True ,
         linewidth = 0.7,
         linestyle = '--',
         color = 'gray')

for bar in ax.patches:
       ax.annotate( s = str(bar.get_height()) , # the text to put
                    xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                    ha='center',
                    va='center',
                    size=14,
                    xytext=(0, 10),
                    textcoords='offset points')

plt.title("Distribution of Educational background of all winners" ,size=19 ,color='pink');
```

```python
df.groupby(['Education']).sum()
```

```python
# Plotting education vs Total assets on pie chart
fig1, ax1 = plt.subplots()
patches, texts, autotexts = ax1.pie(x= df.groupby(['Education']).sum()['Total Assets'] , 
                                    labels=df.groupby(['Education']).sum().index,
                                    autopct='%1.0f%%' ,radius = 3,colors=sns.set_palette('Set1'))
for text in texts:
    text.set_color('pink')
    text.set_size(15)
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_size(15)
```
